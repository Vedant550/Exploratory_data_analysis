## Exploratory Data Analysis

### For Gender

```python
sns.countplot(x='Gender', data = df)
```

```python
ax = sns.countplot(x='Gender', data=df)

for bar in ax.containers:
    ax.bar_label(bar)
```

### To find Out who did the MosT Shopping

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending= False)
```

```python
# plotting a bar chart for gender vs total amount

sales_gender= df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.barplot(x= 'Gender', y = 'Amount', data = sales_gender)
```

 BY THIS WE KNOW, MOST BUYERS ARE FEMALE AND EVEN THE PURCHASING POWER OF WOMEN IS GREATER THAN MEN

### AGE

```python
df.columns
```

```python
sns.countplot(x= 'Age Group',data = df)
```

```python
sns.countplot(data = df, x ='Age Group', hue= 'Gender')
```

```python
ax = sns.countplot(data= df, x='Age Group', hue  ='Gender')

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
### Total Amount Vs Age Group


sales_age= df.groupby('Age Group', as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)


sns.barplot(x='Age Group', y ='Amount', data=sales_age)
```

From Above Graphs, we see Most of the Buyers are from 26-35 Age group and Those are Women

### States

```python
df.columns
```

```python
sns.countplot(x ='State', data =df)

sns.set(rc={'figure.figsize':(25,10)})
```

```python
df.columns
```

```python
### Orders Per state

sales_state= df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending = False)

sns.barplot(x='State', y ='Orders', data = sales_state)
```

```python
sales_state= df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending = False).head()

sns.barplot(x='State', y ='Orders', data = sales_state)
```

```python
### Amount Spent Per state


state_amount= df.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False)


sns.barplot(x ='State', y ='Amount', data = state_amount)
```

From abv Graphs, we see Most of the Orders and Amount Spent is from UP and MAHARASHTRA and KARNATAKA

### Marital Status

```python
df.columns
```

```python
ax = sns.countplot(x='Marital_Status',data=df, hue='Gender')


for bar in ax.containers:
    ax.bar_label(bar)
```

```python
shadi_sales = df.groupby(['Marital_Status','Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False)


sns.barplot(x='Marital_Status', y ='Amount', data= shadi_sales, hue='Gender')
```

From abv Chart, Married Women have the high Purchasing Power

### Occupation

```python
df.columns
```

```python
ax= sns.countplot(x='Occupation', data=df)

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
Ocp_sales = df.groupby(['Occupation'],as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False)


sns.barplot(x='Occupation',y= 'Amount',data =Ocp_sales)
```

From abv : IT sector, HealthCare and Aviation spents the Most Amount

### Product Category

```python
df.columns
```

```python
ax = sns.countplot(x='Product_Category', data=df)


for bar in ax.containers:
    ax.bar_label(bar)
```

```python
prd_sales = df.groupby(['Product_Category', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)


sns.barplot(x='Product_Category', y ='Amount', data= prd_sales, hue='Gender')
```

We observe Most Sold products are from FOOD, FOOTWARE & ELECTRONICS purchased by Womens the most!

### To know, the Top 10 Selling Products

```python
df.columns
```

```python
top_prd = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)


sns.barplot(x='Product_ID', y='Orders', data= top_prd)
```

### So Conclusion,
